```python
import fitz
from pprint import pprint
import pandas as pd


new_indice = ['A','B','C','D','E']
new_columns1 = ['Strongly agree', 'Agree', 'Somewhat agree', 'Somewhat disagree', 'Disagree', 'Strongly disagree', 'Don\'t know']
new_columns2 = ['Strongly agree', 'Agree', 'Somewhat agree', 'Somewhat disagree', 'Disagree', 'Don\'t know']
doc = fitz.open(r"../Sample.pdf",stream=None) # open document
length=len(doc)
print(length)
for i in range(0,length):
   page = doc[i] # get the 1st page of the document
   tabs = page.find_tables() # locate and extract any tables on page
   print(f"{len(tabs.tables)} found on {page}") # display number of found tables
   if tabs.tables:  # at least one table found?.
      df=pd.DataFrame(tabs[0].extract())
      df=df.iloc[:-2,2:-1]
      df=df[df.iloc[:,0]=="Count"]
      df.drop([2],axis=1,inplace=True)
      df.columns=new_columns1
      df.index=new_indice
      df=pd.DataFrame(df.stack(future_stack=False))
      #df.reset_index().melt(id_vars='index')
      flattened_df = df.reset_index()
      flattened_df.columns=['Sepc Group','Response','Count']
      print(flattened_df)
   
```

```python
import pandas as pd

data = {
    'Strongly agree': [10, 100],
    'Agree': [10, 100],
    'Somewhat agree': [10, 100],
    'Somewhat disagree': [10, 100],
    'Disagree': [10, 100],
    'Strongly disagree': [10, 100],
    'Don\'t know': [10, 100]
}
Ind = pd.MultiIndex.from_product([['Asian or Asian American', 'Black or African American'],
                                    ['Strongly agree', 'Agree', 'Somewhat agree', 'Somewhat disagree', 
                                     'Disagree', 'Strongly disagree', 'Don\'t know']],
                                   names=['Group', 'Response'])


data1=pd.DataFrame(data).stack()
print(data1)
df = pd.DataFrame(data=data1,index=Ind)
print(df)

# Flattening the multi-index DataFrame
flattened_df = df.reset_index()

# Display the flattened DataFrame
print(flattened_df)
```
